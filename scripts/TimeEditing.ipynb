```python
import pandas as pd
import datetime
import csv
```

```python
data = pd.read_csv("../data/clean_cab_rides.csv")
```

```python
data = data.drop(columns=['id'])
```

```python
#morning = 05:00:00 - 12
#afternoon = 12:00:00 - 17:00:00
#evening = 17:00:00 - 22:00:00
#night = 22:00:00 - 00:00:00
timeOfDayDF = []
dayOfWeekDF = []
for row in data['time_stamp']:
    #Get time of day
    hour = int(row[11:13])
    if hour >= 5 and hour < 12:
        timeOfDay = "Morning"
    elif hour >= 12 and hour < 17:
        timeOfDay = "Afternoon"
    elif hour >= 17 and hour < 22:
        timeOfDay = "Evening"
    else:
        timeOfDay = "Night"
    timeOfDayDF.append(timeOfDay)
    
    #Get day of week
    year = int(row[0:4])
    month = int(row[5:7])
    day = int(row[8:10])
    date = datetime.datetime(year, month, day)
    dayOfWeek = date.weekday()
    #Convert number to day string
    if dayOfWeek == 0:
        dayOfWeek = "Monday"
    elif dayOfWeek == 1:
        dayOfWeek = "Tuesday"
    elif dayOfWeek == 2:
        dayOfWeek = "Wednesday"
    elif dayOfWeek == 3:
        dayOfWeek = "Thursday"
    elif dayOfWeek == 4:
        dayOfWeek = "Friday"
    elif dayOfWeek == 5:
        dayOfWeek = "Saturday"
    else:
        dayOfWeek = "Sunday"
    dayOfWeekDF.append(dayOfWeek)
    
timeOfDayDF = pd.DataFrame(timeOfDayDF, columns=['Time Of Day'])
dayOfWeekDF = pd.DataFrame(dayOfWeekDF, columns=['Day of Week'])
```

```python
result = dayOfWeekDF.join(data)
result = timeOfDayDF.join(result)
result.head()
```

```python
result.to_csv('../data/clean_cab_rides.csv', index=False, quoting=csv.QUOTE_NONNUMERIC)
```
